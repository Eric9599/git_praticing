```python
import time
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel
```

```python
# 1. æ¨¡å‹åˆå§‹åŒ– (temperature=0, ä½¿ç”¨ä½ çš„é…ç½®)
llm = ChatOpenAI(
    base_url="https://ws-02.wade0426.me/v1",
    api_key="day2hw",
    model="Qwen/Qwen3-VL-8B-Instruct",
    temperature=0,
    max_tokens=50
)

# 2. å®šç¾©å…©å€‹ä¸åŒé¢¨æ ¼çš„ Chain (åˆ©ç”¨ LCEL)
prompt1 = ChatPromptTemplate.from_messages([("human", "è«‹ç”¨ã€å¹½é»˜é¢¨ã€‘é‡å°ä¸»é¡Œ {topic} å¯«ä¸€ç¯‡è²¼æ–‡")])
prompt2 = ChatPromptTemplate.from_messages([("human", "è«‹ç”¨ã€å°ˆæ¥­é¢¨ã€‘é‡å°ä¸»é¡Œ {topic} å¯«ä¸€ç¯‡è²¼æ–‡")])

chain1 = prompt1 | llm | StrOutputParser()
chain2 = prompt2 | llm | StrOutputParser()

# 3. å»ºç«‹ã€Œå¤šå·¥è™•ç†ã€AI Agent
parallel_agent = RunnableParallel(style_humor=chain1, style_professional=chain2)

target_topic = {"topic": "é‡å­é›»è…¦"}

# 4. åŸ·è¡Œæ–¹å¼ä¸€ï¼šæµå¼ (Streaming)
print("="*20, "1. Streaming æ¨¡å¼é–‹å§‹", "="*20)
for chunk in parallel_agent.stream(target_topic):
    print(chunk, end="", flush=True)
print("\n")

# 5. åŸ·è¡Œæ–¹å¼äºŒï¼šæ‰¹æ¬¡ (Batch) ä¸¦ç´€éŒ„æ™‚é–“
print("-" * 50) # é€™æ˜¯ä¿®æ­£å¾Œçš„ç·šæ¢ï¼Œæ”¾åœ¨ print è£¡é¢
print("="*20, "2. Batch æ¨¡å¼é–‹å§‹", "="*20)

start_time = time.perf_counter()
batch_result = parallel_agent.invoke(target_topic)
end_time = time.perf_counter()

print("æ‰¹æ¬¡è™•ç†çµæœï¼š", batch_result)
```

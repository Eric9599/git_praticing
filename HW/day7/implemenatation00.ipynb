```python
import pdfplumber
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions
from docling.document_converter import DocumentConverter, PdfFormatOption
from markitdown import MarkItDown
```

```python
def pdf_to_markdown_plumber(pdf_path, output_path):
    with pdfplumber.open(pdf_path) as pdf:
        full_text = ""
        for page in pdf.pages:
            # 提取文字並在每頁結尾加換行
            full_text += page.extract_text() + "\n\n"

    with open(output_path, "w", encoding="utf-8") as f:
        f.write(full_text)

# 使用範例
pdf_to_markdown_plumber("CW/05/example.pdf", "output_plumber.md")
```

```python
def pdf_to_markdown_docling(pdf_path, output_path):
    converter = DocumentConverter()
    result = converter.convert(pdf_path)

    # 直接匯出為 markdown 格式
    markdown_content = result.document.export_to_markdown()

    with open(output_path, "w", encoding="utf-8") as f:
        f.write(markdown_content)

# 使用範例
pdf_to_markdown_docling("CW/05/example.pdf", "output_docling.md")
```

```python
def pdf_to_markdown_microsoft(pdf_path, output_path):
    md = MarkItDown()
    result = md.convert(pdf_path)

    with open(output_path, "w", encoding="utf-8") as f:
        f.write(result.text_content)

# 使用範例
pdf_to_markdown_microsoft("CW/05/example.pdf", "output_markitdown.md")
```
